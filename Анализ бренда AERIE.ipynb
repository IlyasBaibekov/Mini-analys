```python
df_new = df.dropna(subset=['rating'])
```

```python
dfn = df_new.sort_values(by='price')
dfn.tail(10)
```

```python
df_nnew = dfn[dfn['review_count'] >= 2.0]
df_nnew.head(20)
```

```python
def getmid(dataframe, a, b):
    # Фильтрация строк с ценами в диапазоне от a до b
    s = dataframe[(dataframe['price'] >= a) & (dataframe['price'] <= b)]
    # Вычисление среднего рейтинга
    return s['rating'].mean()


print(getmid(df_nnew,3.0, 10.0))
print(getmid(df_nnew,10.01, 17.0))
print(getmid(df_nnew,17.01, 24.0))
print(getmid(df_nnew,24.01, 31.0))
print(getmid(df_nnew,31.01, 40.0))
```

```python
print(getmid(dfn,3.0, 10.0))
print(getmid(dfn,10.01, 17.0))
print(getmid(dfn,17.01, 24.0))
print(getmid(dfn,24.01, 31.0))
print(getmid(dfn,31.01, 45.0))
```

```python
import matplotlib.pyplot as plt

# Данные для диаграммы
ratings = [
    4.355254988913526,
    4.39741870600067,
    4.351099384344767,
    3.1433447098976104,
    3.8879518072289163
]

# Создание списка меток для столбцов
labels = [f'Item {i+1}' for i in range(len(ratings))]

# Создание столбчатой диаграммы
plt.figure(figsize=(10, 6))
plt.bar(labels, ratings, color='skyblue')

# Добавление заголовков и подписей
plt.xlabel('Items')
plt.ylabel('Rating')
plt.title('Ratings for Items')
plt.ylim(0, max(ratings) + 1)  # Установка предела по оси Y для лучшего отображения

# Показ диаграммы
plt.show()
```
